```python
# Car Price Prediction
#Linear Regression vs Random Forest

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
 ## 1) Load the cleaned dataset
df = pd.read_csv("/content/drive/MyDrive/assignmentFour/car_dataset.csv")
print(df.head())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove index column if it exists
df = df.drop(columns=["Unnamed: 0"])
```

```python
# 2. Prepare Features & Target
# Target Variable
y = df["Price"]

# Feature Variables
X = df.drop(columns=["Price", "LogPrice"])

print(X.head())
print(y.head())
```

```python
# 3. Split Data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape :", X_test.shape)
```

```python
# 4. Train Models
#Linear Regression
lr = LinearRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

#Random Forest
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)


```

```python
# 5. Evaluate Performance

def print_metrics(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    print(f"\n{name} Performance")
    print(f"R²   : {r2:.3f}")
    print(f"MAE  : {mae:.2f}")
    print(f"MSE  : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")

# Call the function
print_metrics("Linear Regression", y_test, lr_pred)
print_metrics("Random Forest", y_test, rf_pred)
```

```python
#6. Sanity Check
i = 3

sample = X_test.iloc[[i]]
actual_price = y_test.iloc[i]

linear_prediction = lr.predict(sample)[0]

forest_prediction = rf.predict(sample)[0]

print("Actual Price:", actual_price)
print("Linear Regression Prediction:", linear_prediction)
print("Random Forest Prediction:", forest_prediction)
```
